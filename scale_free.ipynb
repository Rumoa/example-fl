```python
def compute_largest_mse_mean(data):
    data_arr = np.array(data)
    mse_per_eigval_real = np.mean(np.real((data_arr[ :, 0, :] - data_arr[ :, 1, :]))**2, axis=0)
    mse_per_eigval_imag = np.mean(np.imag((data_arr[ :, 0, :] - data_arr[ :, 1, :]))**2, axis=0)
    return mse_per_eigval_real[-1], mse_per_eigval_imag[-1]

def compute_largest_mse(data):
    data_arr = np.array(data)
    mse_per_eigval_real = np.real((data_arr[ :, 0, :] - data_arr[ :, 1, :]))**2
    mse_per_eigval_imag = np.imag((data_arr[ :, 0, :] - data_arr[ :, 1, :]))**2
    return mse_per_eigval_real[:, -1], mse_per_eigval_imag[:, -1]    
```

```python
mean_unravelled_real = []
mean_unravelled_imag = []
std_unravelled_real = []
std_unravelled_imag = []

for i in mean_std_perdim:
    mean_unravelled_real.append(i[0,0])
    mean_unravelled_imag.append(i[1,0])
    std_unravelled_real.append(i[0,1])
    std_unravelled_imag.append(i[1,1])

mean_unravelled_real = np.array(mean_unravelled_real)
mean_unravelled_imag = np.array(mean_unravelled_imag)
std_unravelled_real = np.array(std_unravelled_real )
std_unravelled_imag =  np.array(std_unravelled_imag )   
```
